# Cross-matching Swift and LaMassa

We have produced a catalog from the Swift DeepSky pipeline, where LaMassa's Chandra pointings were used to select the Swift observations.
Now we want to take the results from such processing and correlate with LaMassa's.

```python
#swift_file = 'table_flux_detections_swift_lamassa_chandra_unique.csv'
swift_file = 'swift_deepsky/table_flux_detections_stripe82_unique.csv'
lamassa_file = 'lamassa/chandra.dat'
```

#### Data

```python
# Target catalog
#
import pandas as pd
sds = pd.read_csv(swift_file, sep=';')

# Ancillary catalog
#
from astropy.table import Table
lamassa = Table.read(lamassa_file, format='cds', readme='lamassa/ReadMe')
```

```python
lamassa[:5]
```

```python
sds[:5]
```

```python
# Let's convert SDS' coordinates to degrees, and check the wrap up at 180deg since we're covering -60<RA<60
#
from astropy.coordinates import SkyCoord
from astropy import units

# swift_coords = SkyCoord(sds['RA'], sds['DEC'], unit=(units.hourangle,units.degree))
swift_coords = SkyCoord(sds['RA'], sds['DEC'], unit=(units.degree,units.degree))

sds['RA'] = swift_coords.ra.wrap_at('180d').deg
sds['DEC'] = swift_coords.dec.deg

sds.head()
```

```python
# Convert to pandas df and wrap the sky also:
#
coords = SkyCoord(lamassa['RAdeg'], lamassa['DEdeg'], unit=(units.degree,units.degree))
lamassa['RAdeg'] = coords.ra.wrap_at('180d')
lamassa['DEdeg'] = coords.dec

lamassa = lamassa.to_pandas()
```

## xmatch the neighbourhood

Well, we'll use `xmatch` to search for targets' counterparts, the default is to search for the nearest neighbour within a searching radius.

We may use the average positional error from the catalog(s) as the search `radius`.

```python
from astropy import units
rs = lamassa['e_Pos'].mean() * units.arcsec
```

```python
from xmatch import xmatch
help(xmatch)
```

```python
colsA = dict(ra='RAdeg', dec='DEdeg', id='MSID')
colsB = dict(ra='RA', dec='DEC', id='OBJID')

xcat = xmatch(lamassa, sds, columns_A=colsA, columns_B=colsB, radius=rs)
```

```python
inds = xcat[('AB','separation')].dropna().index
xcat = xcat.iloc[inds]
```

```python
s_idx = xcat[('B','OBJID')].astype(int)
matched_swift = sds.set_index('OBJID').loc[s_idx]
matched_swift.head(10)
```

```python
matched_swift.reset_index(inplace=True)
```

```python
s_idx = xcat[('A','MSID')].astype(int)
matched_lamassa = lamassa.set_index('MSID').loc[s_idx]
matched_lamassa.head(10)
```

```python
matched_lamassa.reset_index(inplace=True)
```
