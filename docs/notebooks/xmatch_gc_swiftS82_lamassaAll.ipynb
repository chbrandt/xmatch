## xmatch using 'gc' algorithm

Fot a great-circle matching we have to define the search radius.
I will use LaMassa's mean positional error for that.

```python
lm_chandra = lm_chandra.to_pandas()
```

```python
from astropy import units

rs = lm_chandra['e_Pos'].mean() * units.arcsec
```

```python
from xmatch import xmatch

help(xmatch)
```

```python
colsA = dict(ra='RAdeg', dec='DEdeg', id='MSID')
colsB = dict(ra='RA', dec='DEC', id='OBJID')

xcat = xmatch(lm_chandra, swift, columns_A=colsA, columns_B=colsB, radius=rs)
```

```python
xcat
```

```python
inds = xcat[('AB','separation')].dropna().index
xcat = xcat.iloc[inds]
```

```python
s_idx = xcat[('B','OBJID')].astype(swift['OBJID'].dtype)
matched_swift_chandra = swift.set_index('OBJID').loc[s_idx]
matched_swift_chandra
```

```python
matched_swift_chandra.reset_index(inplace=True)
```

```python
s_idx = xcat[('A','MSID')].astype(lm_chandra['MSID'].dtype)
matched_lm_chandra = lm_chandra.set_index('MSID').loc[s_idx]
matched_lm_chandra
```

```python
matched_lm_chandra.reset_index(inplace=True)
```

```python
len(matched_lm_chandra) == len(matched_swift_chandra)
```

```python
del lm_chandra, xcat
```

### xmatch LaMassa-xmm10

```python
lm_xmm10 = lm_xmm10.to_pandas()
lm_xmm10.sample(10)
```

```python
from astropy import units
rs = lm_xmm10['e_Pos'].mean() * units.arcsec


from xmatch import xmatch
colsA = dict(ra='RAdeg', dec='DEdeg', id='Seq')
colsB = dict(ra='RA', dec='DEC', id='OBJID')

xcat = xmatch(lm_xmm10, swift, columns_A=colsA, columns_B=colsB, radius=rs)

inds = xcat[('AB','separation')].dropna().index
xcat = xcat.iloc[inds]

s_idx = xcat[('B','OBJID')].astype(swift['OBJID'].dtype)
matched_swift_xmm10 = swift.set_index('OBJID').loc[s_idx]
matched_swift_xmm10.reset_index(inplace=True)

s_idx = xcat[('A','Seq')].astype(lm_xmm10['Seq'].dtype)
matched_lm_xmm10 = lm_xmm10.set_index('Seq').loc[s_idx]
matched_lm_xmm10.reset_index(inplace=True)

len(matched_lm_xmm10) == len(matched_swift_xmm10)
```

```python
matched_lm_xmm10
```
